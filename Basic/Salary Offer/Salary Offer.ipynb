This is a basic linear regreesion model which predict the supposed salary offered to client. This use the formula **Y = mx + b**.

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from word2number import w2n

from google.colab import files
files.upload()
df = pd.read_csv('hiring.csv')
df
```

```python
df.experience = df.experience.fillna('zero')
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
import math
median_score = math.floor(df['test_score(out of 10)'].mean())
median_score
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_score)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```
